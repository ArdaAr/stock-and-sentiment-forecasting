```python
# WORK
dates = []
authors = []
content = []
main_links = []
# Perulangan tiap Halaman

page_start = 67
page_end = 85 # 67 + 1


for page in range(page_start,page_end):    
    url='https://search.bisnis.com/?q=bca&per_page='+str(page)
#     print(2)
    r = requests.get(url)
    b = BeautifulSoup(r.content,'lxml')

    # get Links from page
    links = []
    for news in b.find_all('h2')[0:10]:
#         print(3)
        links.append(news.a['href'])
    
    # kode dibawah digunakan untuk mengambil paragraf pertama dari isi setiap berita
    for link in links:
#         print(1)
        halaman = requests.get(link)
        bsobj = BeautifulSoup(halaman.content,'lxml')
        try :
            matches = ['BCA','bca','Bank Central Asia']
            isi = ''
            for news in bsobj.find('div',{'class':'col-sm-10'}).findAll('p'): # mendapat statement BCA dari berita
                if any(x in news.text for x in matches):
                    isi += news.text
#                     print('#')
#                     print(news.text)
#             text = bsobj.find('p').text
            date = bsobj.find('div',{'class':'new-description'}).find('span').text.strip()
            author = bsobj.find('div',{'class':'new-description'}).find('a').text.strip()
#             print('#'*20)
#             print(text)
            authors.append(author)
            dates.append(date)
            content.append(isi)
            main_links.append(link)
        except:
            continue
```

```python
df_news = pd.DataFrame({
    'date' : dates,
    'author' : authors,
    'news' : content,
    'link' : main_links
})
```

# Scrapping from kontan.co.id

```python
# WORK
dates = []
content = []
main_links = []
headlines = []
# Perulangan tiap Halaman

page_start = 300
page_end = 381 # 380 + 1


for page in range(page_start,page_end):
    url='https://www.kontan.co.id/search/?search=BBCA&per_page='+str(page)
#     print(2)
    r = requests.get(url)
    b = BeautifulSoup(r.content,'lxml')

    # get Links from page
    links = []
    for news in b.findAll('div',{'class':'sp-hl linkto-black'}):
#         print(3)
        links.append('https:'+news.a['href'])
    
    # kode dibawah digunakan untuk mengambil paragraf pertama dari isi setiap berita
    for link in links:
#         print(1)
        halaman = requests.get(link)
        bsobj = BeautifulSoup(halaman.content,'lxml')
        matches = ['BBCA','BCA','bca','Bank Central Asia']
        isi = ''
        try:
            judul = bsobj.find('h1').text
            tanggal_berita = bsobj.find('div',{'class':'fs14 ff-opensans font-gray'}).text.strip()
            for news in bsobj.find('div',{'class':'tmpt-desk-kon'}).findAll('p'):
                print(3) # mendapat statement BCA dari berita
                if any(x in news.text for x in matches):
                    isi += news.text
            headlines.append(judul)
            dates.append(tanggal_berita)
            content.append(isi)
            main_links.append(link)
        except:
            continue
```

```python
df_news = pd.DataFrame({
    'date' : dates,
    'headline' : headlines,
    'news' : content,
    'link' : main_links
})
```

```python
# mengambil string tanggal lalu assign ke kolom baru 'datetime'

df_news['datetime'] = df_news['date'].apply(lambda x: x.rpartition('|')[0].rstrip())
df_news
```

```python
df_news['datetime'][1].replace(' ','-')
```

```python
# mengubah bulan string menjadi angka

def change_month(date):
    date = date.replace(' ','-')
    if 'Juli' in date:
        return date.replace("Juli", '07')
    elif 'Juni' in date:
        return date.replace("Juni", '06')
    elif 'Mei' in date:
        return date.replace("Mei", '05')
    elif 'April' in date:
        return date.replace("April", '04')

df_news['datetime'] = df_news['datetime'].apply(lambda x: change_month(x))
df_news
```

```python
# mengubah format tanggal

df_news['datetime'] = pd.to_datetime(df_news['datetime'],format='%d-%m-%Y')
df_news
```

# Natural Language Processing

```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from nltk import word_tokenize, sent_tokenize
from nltk import FreqDist
from nltk.corpus import stopwords
from wordcloud import WordCloud, ImageColorGenerator
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from nltk.stem import WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
ind = stopwords.words('indonesian')
```

```python
df_berita = df_news[['news','sentiment']]
df_processed = df_berita.copy()
df_processed.head()
```

```python
# periksa null
df_processed.isnull().sum()
```

```python
df_processed.dropna(inplace=True)
df_processed.isnull().sum()
```

```python
# mengubah text ke lower

df_processed['news'] = df_processed['news'].apply(lambda x : x.lower())
df_processed.head()
```

```python
# Hapus singkatan ?
...
```

```python
# mengubah kalimat menjadi kata

df_processed['news'] = df_processed['news'].apply(lambda x : word_tokenize(x))
df_processed['news']
```

```python
# Hapus stopwords indonesia

def removeStopWords(text):
    words = []
    for word in text:
        if word not in ind:
            words.append(word)
    return words

df_processed['news'] = df_processed['news'].apply(lambda x : removeStopWords(x))
df_processed['news']
```

```python
# Hapus simbol " (), '', "

def hapusSimbol(sentence): #tokenize first
    words = []
    for word in sentence:
        if word.isalpha():
            words.append(word)
    return words

df_processed['news'] = df_processed['news'].apply(lambda x : hapusSimbol(x))
df_processed['news']
```

```python
# Stemming

lemmatizer_obj = WordNetLemmatizer()
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stemm(data):
    words = []
    for word in data:
        stm = stemmer.stem(word)
        words.append(stm)
    return words

df_processed['news'] = df_processed['news'].apply(lambda x: [stemmer.stem(y) for y in x])
df_processed['news']
```

```python
# mengubah list menjadi set

df_processed['news'] = [ set(w) for w in df_processed['news']]
df_processed['news']
```

```python
# memeriksa data kosong akibat cleaning
dataKosong = 0
for i in df_processed['news']:
    if len(i)==0:
        dataKosong +=1
print(dataKosong)
```

```python
# mengubah dari set/list ke text

df_processed['news'] = [ " ".join(data) for data in df_processed['news']]
df_processed['news']
```
