# Checkpoint

- Scrapping berdasarkan hashtag atau topik tidak dapat di ambil berdasarkan tanggal spesifik, data yang diambil adalah data hari dimana kita melakukan scrapping

- Scrapping data berdasarkan user dapat di spesifikan sesuai range tanggal yang kita tentukan
- Bisa scrapping sesuai tanggal, namun fungsi tidak bisa mendefinisikan bahasa tweet, sehingga masih tercampur dengan bahasa asing
- Scrapping dengan tanggal spesifik sudah berhasil menggunakan library snscrape
- agar data yang diambil sesuai dengan range tanggal yang didefinisikan, penulisan format tanggal harus sampai detik 

# Scrapping data train tweets

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for i,tweet in enumerate(sntwitter.TwitterSearchScraper('BCA').get_items()):
    if i>1000:
        break
    tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df3 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```

```python
tweets_df3.to_csv('../Code 2/data/tweet_mbankingbca.csv',index=False)
```

# Srapping tweets for testing

```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

# Creating list to append tweet data to
tweets_list2 = []

# Using TwitterSearchScraper to scrape data and append tweets to list
for day in range(3,17):
    for i,tweet in enumerate(sntwitter.TwitterSearchScraper(f'mbankingbca until:2022-04-0{day} lang:id').get_items()):
        if i>50:
            break
        tweets_list2.append([tweet.date, tweet.id, tweet.content, tweet.user.username])
    
# Creating a dataframe from the tweets list above
tweets_df2 = pd.DataFrame(tweets_list2, columns=['Datetime', 'Tweet Id', 'Text', 'Username'])
```
